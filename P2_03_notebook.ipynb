### <span style='color:orange'> Creation de l'indicateur composite </span> 

L'indicateur composite est une somme pondérée des différents indicateurs. Les coefficients de pondération doivent suivre le niveau d'importance des variables. On part sur une pondération en base 1 (en pourcentage)

  <span style='color:green'> 1- GDP per capita, PPP (current international §) -  % 15 </span> 

<span style='color:green'> 2- Government expenditure on secondary education as % of GDP (%)' - % 10  </span> 

<span style='color:green'>  3- Government expenditure on tertiary education as % of GDP (%)'-  % 10  </span> 

<span style='color:green'> 4- Gross enrolment ratio, tertiary, both sexes (%)' - % 10 </span> 

<span style='color:green'> 5 - Adjusted net enrolment rate, secondary, both sexes (%)'] - % 15  </span> 

<span style='color:green'> 6- Internet users (per 100 people)' - % 20  </span> 

<span style='color:green'> 7- Population, ages 15-24, total' - % 20  </span> 

On va remplacer les NaN par 0

```python
df_short_indicateur_Pivot['Indicateur_composite']=(df_short_indicateur_Pivot['GDP per capita, PPP (current international $)']*0.15 + df_short_indicateur_Pivot['Government expenditure on secondary education as % of GDP (%)']*0.10 + df_short_indicateur_Pivot['Government expenditure on tertiary education as % of GDP (%)']*0.10 + df_short_indicateur_Pivot['Gross enrolment ratio, tertiary, both sexes (%)']*0.10 + df_short_indicateur_Pivot['Population, ages 15-24, total']*0.20 +  df_short_indicateur_Pivot['Internet users (per 100 people)']*0.20+  df_short_indicateur_Pivot['Adjusted net enrolment rate,  secondary, both sexes (%)']*0.15)  
```

```python
df_short_indicateur_Pivot = df_short_indicateur_Pivot.replace([np.nan], 0)
```

```python
minmax_scale100 = MinMaxScaler((0,100))
df_short_indicateur_Pivot['Indicateur_composite']=minmax_scale100.fit_transform(np.array(df_short_indicateur_Pivot['Indicateur_composite']).reshape(-1, 1))
```
